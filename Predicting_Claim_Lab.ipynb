```python
def log_transform_clean(x):
    x = np.log(x)
    if np.isfinite(x): 
        return x
    else:
        return np.NAN
    
data_num['customer_lifetime_value_log']=list(map(log_transform_clean, data_num['customer_lifetime_value']))
```

Encoding Categorical Data fields using OHE.

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(data_cat)
encoded = encoder.transform(data_cat).toarray()
encoded
```

```python
data_cat = encoded
```

Bring categorical and numerical columns back together using pd.concat.

Bring categorical and numerical columns back together using pd.concat.

```python
X = np.concatenate((data_num, data_cat), axis=1)
```

```python
y = data['total_claim_amount']
```

```python
X.shape
```

```python
y.shape
```

Splitting into train set and test dataset using random state, eg 80%:20% .

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=20)
```

# 05 - Modeling
Apply linear regression model from sklearn.linear_model.

```python
lm= linear_model.LinearRegression()
model=lm.fit(X_train, y_train)
predictions=lm.predict(X_test)
r2_score(y_test, predictions)
```
